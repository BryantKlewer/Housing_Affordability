```python
import pandas as pd
import numpy as np
import numpy_financial as npf
```

```python
interest_data = pd.read_csv("Resources\Freddie_Mac_15_Year_30_Year_Interest_Rate.csv")
home_data = pd.read_csv("Resources\Zillow_Single_Family_Home_Index_by_Metro_US_Consolidated.csv")
tax_data = pd.read_csv("Resources\Aggregated_Median_Tax_and_Monthly_Cost.csv")
```

```python
clean = {'Owner_occupied_housing_units_with_a_mortgage': str,
                'Median_Taxes': str
                }
tax_data = tax_data.astype(clean)
tax_data['Owner_occupied_housing_units_with_a_mortgage']=tax_data['Owner_occupied_housing_units_with_a_mortgage'].str.replace(',','')
```

```python
tax_data['Median_Taxes']=tax_data['Median_Taxes'].str.replace(',','')
tax_data.head()
```

```python
tax_data[['Owner_occupied_housing_units_with_a_mortgage', 'Median_Taxes']] = tax_data[['Owner_occupied_housing_units_with_a_mortgage', 'Median_Taxes']].apply(pd.to_numeric)
tax_data.dtypes
```

```python
tax_data_df= tax_data.sort_values(by=['City_Rank_by_Population(2021)', 'Observation_Date'])
tax_data_df.head()
```

```python
tax_data_df.shape
```

```python
combo_df = tax_data_df.merge(home_data_df, on =['City_Rank_by_Population(2021)', 'Observation_Date'], how='left', indicator = True)
combo_df.head()
```

```python
combo_df.shape
```

```python
combo_home_tax_df = combo_df.drop(combo_df.columns[[5, 7, 8, 9, 10, 12, 13]], axis=1)
combo_home_tax_df = combo_home_tax_df.rename(columns={'State_x': 'State', 'State_Abbrieviation_x': 'State_Abbrieviation', 'City_x': 'City',})
combo_home_tax_df.head()
```

```python
home_tax_interest_df = combo_home_tax_df.merge(interest_data, on = 'Observation_Date', how='left', indicator = True)
home_tax_interest_df.head()
```

```python
# BASE DF TO BUILD ONTO
home_tax_interest_df = home_tax_interest_df.drop(home_tax_interest_df.columns[[8, 10, 11, 12]], axis=1)
home_tax_interest_df.head()
```

```python
home_tax_interest_df = home_tax_interest_df.rename(columns={'15_Year_Annual_Average_Rate': 'AnnualAverageRate15Year', '30_Year_Annual_Average_Rate': 'AnnualAverageRate30Year' })
home_tax_interest_df.head()
```

```python
# 15yr mortgage df 
fifteen_year_df = home_tax_interest_df.drop('AnnualAverageRate30Year', axis=1)
fifteen_year_df.head()
```

```python
# calculate payments for 10% down and 15yr mortgage
# npf.pmt(rates/1200, term *12, -loan_amount)
def PaymentFunction(row):
    x = npf.pmt(row.AnnualAverageRate15Year/1200, 180, row.Median_Typical_Home_Value*-.9)
    return x

home_tax_interest_df["15yr_Payment_10_Perc_Down"] = fifteen_year_df.apply(PaymentFunction, axis = 1)
home_tax_interest_df.head()
```

```python
# calculate payments for 20% down and 15yr mortgage
def PaymentFunction1(row):
    x = npf.pmt(row.AnnualAverageRate15Year/1200, 180, row.Median_Typical_Home_Value*-.8)
    return x

home_tax_interest_df["15yr_Payment_20_Perc_Down"] = fifteen_year_df.apply(PaymentFunction1, axis = 1)
home_tax_interest_df.head()
```

```python
# 30yr mortgage df 
thirty_year_df = home_tax_interest_df.drop('AnnualAverageRate15Year', axis=1)
thirty_year_df.head()
```

```python
# calculate payments for 10% down and 30yr mortgage
# npf.pmt(rates/1200, term *12, -loan_amount)
def PaymentFunction2(row):
    x = npf.pmt(row.AnnualAverageRate30Year/1200, 360, row.Median_Typical_Home_Value*-.9)
    return x

home_tax_interest_df["30yr_Payment_10_Perc_Down"] = thirty_year_df.apply(PaymentFunction2, axis = 1)
home_tax_interest_df.head()
```

```python
# calculate payments for 20% down and 30yr mortgage
def PaymentFunction3(row):
    x = npf.pmt(row.AnnualAverageRate30Year/1200, 360, row.Median_Typical_Home_Value*-.8)
    return x

home_tax_interest_df["30yr_Payment_20_Perc_Down"] = thirty_year_df.apply(PaymentFunction3, axis = 1)
home_tax_interest_df.head()
```

```python
completed_df = home_tax_interest_df[['City_Rank_by_Population(2021)', 'State', 'State_Abbrieviation', 'City', 'Observation_Date', 'Median_Taxes', 'Median_Typical_Home_Value', 'AnnualAverageRate15Year', '15yr_Payment_10_Perc_Down', '15yr_Payment_20_Perc_Down', 'AnnualAverageRate30Year', '30yr_Payment_10_Perc_Down', '30yr_Payment_20_Perc_Down']]
completed_df.head()
```

```python
completed_df['15yr_Payment_10_Perc_Down'] = complete_df['15yr_Payment_10_Perc_Down'].apply(lambda x: np.round(x,decimals=2))
completed_df.head()
```

```python
completed_df['15yr_Payment_20_Perc_Down'] = complete_df['15yr_Payment_20_Perc_Down'].apply(lambda x: np.round(x,decimals=2))
completed_df.head()
```

```python
completed_df['30yr_Payment_10_Perc_Down'] = complete_df['30yr_Payment_10_Perc_Down'].apply(lambda x: np.round(x,decimals=2))
completed_df.head()
```

```python
completed_df['30yr_Payment_20_Perc_Down'] = complete_df['30yr_Payment_20_Perc_Down'].apply(lambda x: np.round(x,decimals=2))
completed_df.head()
```
