```python
# import linregress
from scipy.stats import linregress

# create a function to perform linear regression on the home value and income data and plot regression line and equation
def plot_linear_regression(x_values, y_values, title, y_label, text_coordinates):
    
    # run regression on data
    (slope, intercept, r_value, p_value, std_err) = linregress(x_values, y_values)
    
    # calculate the regression line "y values" from slope and intercept
    regress_values = x_values * slope + intercept
    
    # get equation of line
    line_eq = "y = " + str(round(slope,2)) + "x = " + str(round(intercept,2))
    
     # create a scatter plot and plot the regression line
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r")
    # text for the line equation
    plt.annotate(line_eq, text_coordinates, fontsize=15, color="red")
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel(y_label)
    plt.show()
```

```python
x_values=df01["Identifier"]
y_values=df01["Single_Family_Median_Typical_Home_Value"]
plot_linear_regression(x_values, y_values, 'Linear Regression of Single Family Home Values', 'Home Value', (201400,450000))
```

```python
x_values=df01["Identifier"]
y_values=df01["Estimated_Median_Household_Income"]
plot_linear_regression(x_values, y_values, 'Linear Regression of Median Household Income', 'Price', (201500,51000))
```
